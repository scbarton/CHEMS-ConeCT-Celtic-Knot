```python
def rotate(p,angle):
    R=array([[cos(angle),-sin(angle)],[sin(angle),cos(angle)]])
    return inner(R,p).T
```

```python
pa1=array([[-0.45, -0.25], [-0.5 , -0.3 ], [-0.5 , -0.5 ], [-0.3 , -0.5 ], [-0.05, -0.25]])
pb1=array([[-0.15, -0.45], [-0.1 , -0.5 ], [ 0.1 , -0.5 ], [ 0.35, -0.25]])
pc1=array([[ 0.05, -0.15],[0.35, 0.15]]) # .35, .15
```

```python
pa2=rotate(pa1,pi/2)
pa3=rotate(pa1,pi)
pa4=rotate(pa1,3*pi/2)
```

```python
pb2=rotate(pb1,pi/2)
pb3=rotate(pb1,pi)
pb4=rotate(pb1,3*pi/2)
```

```python
pc2=rotate(pc1,pi/2)
pc3=rotate(pc1,pi)
pc4=rotate(pc1,3*pi/2)
```

```python
kwargs={'fc':'none', 'lw':22, 'capstyle':'butt', 'joinstyle':'round', 'closed':False}
targs={'fontname':'impact', 'size':88}

with xkcd(scale=3,length=400, randomness=10):

    fill(*pa1.T, ec='#59004C', **kwargs)
    fill(*pa2.T, ec='#13362C', **kwargs)
    fill(*pa3.T, ec='#59004C', **kwargs)
    fill(*pa4.T, ec='#13362C', **kwargs)

    fill(*pb1.T, ec='#E36C1A', **kwargs)
    fill(*pb2.T, ec='#E36C1A', **kwargs)
    fill(*pb3.T, ec='#E36C1A', **kwargs)
    fill(*pb4.T, ec='#E36C1A', **kwargs)

    fill(*pc1.T, ec='#59004C', **kwargs)
    fill(*pc2.T, ec='#13362C', **kwargs)
    fill(*pc3.T, ec='#59004C', **kwargs)
    fill(*pc4.T, ec='#13362C', **kwargs)

    axis('square')
    xlim(-0.6,0.6)
    ylim(xlim())
    axis('off')
    #title('CHEMS', **targs)
    text (-0.57,0.62,'CHE', c='#59004C', **targs)
    text (0.074,0.62,'MS', c='#13362C', **targs)
    text (-0.65,-0.95,'CON', c='#E36C1A', **targs)
    text (0.04,-0.95,'e', c='#6C6C6C', **targs)
    text (0.25,-0.95,'CT', c='#E36C1A', **targs)
    show()
```
